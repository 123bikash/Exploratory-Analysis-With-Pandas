Cleaning the data

Only want USA races of 50km or 50mi, 2020

```python
df[df["Event distance/length"] == "50mi"]
```

Combine 50kn/50mi with isin

```python
df[(df["Event distance/length"].isin(["50km","50mi"]))   &  (df['Year of event'] == 2020)]
```

```python
df[df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA']
```

Combine all the filters together

```python
df[(df["Event distance/length"].isin(["50km","50mi"]))   &  (df['Year of event'] == 2020)  & (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA')]
```

```python
df2 = df[(df["Event distance/length"].isin(["50km","50mi"]))   &  (df['Year of event'] == 2020)  & (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA')]
```

Remove USA from the event name as all the records are of uSA   

```python
df2['Event name'] = df2['Event name'].str.split('(').str.get(0)
```

```python
df2.head()
```

Clean up athlete age

```python
df2['athlete_age'] = 2020-df2['Athlete year of birth']
```

Remove h from athlete performance

```python
df2['Athlete performance'] = df2["Athlete performance"].str.split(' ').str.get(0)
```

```python
df2.head()
```

Dropping columns : Athlete Club , Athlete Country , Athlete year of birth , Athlete Age Category

```python
df2 = df2.drop(['Athlete club' , 'Athlete country' , 'Athlete year of birth' , 'Athlete age category'] , axis = 1)
```

```python
df2.head()
```

Clean up null value

```python
df2.isna().sum()
```

```python
df2 = df2.dropna()
```

Rename columns

```python
df2 = df2.rename(columns= {
    "Year of event" : "year",
    "Event dates" : "date",
    "Event name" : "event_name",
    "Event distance/length" : "distance" , 
    "Event number of finishers" : "total_finishers" ,
    "Athlete performance" : "performance" , 
    "Athlete gender" : "gender" ,
    "Athlete average speed" : "average_speed" ,
    "Athlete ID" : "athlete_id" , 
    "athlete_age" : "athlete_age"
})
```

Questions I want to find out from the data

1. Difference in speed for 50km , 50mi male to female  

```python
df3.groupby(['distance' , 'gender'])['average_speed'].mean()
```

2. What age grouup are the best i 50mi Race (20+ race min)

```python
df3.query('distance == "50mi"').groupby('athlete_age')['average_speed'].agg(['mean' , 'count']).sort_values('mean', ascending = False).query('count>19')
```

3. What age grouup are the best i 50mi Race (20+ race min)

```python
df3.query('distance == "50mi"').groupby('athlete_age')['average_speed'].agg(['mean' , 'count']).sort_values('mean', ascending = True).query('count>19')
```

Are runners lower in summer than winter?
spring (3-5) , 
summer (6-8) ,
fall (9-11) ,
winter (12-2)

```python
df3['race_month'] = df3['date'].str.split('.').str.get(1).astype(int)
```

```python
df3['race_season'] = df3['race_month'].apply(lambda x: 'Winter' if x>11 else 'Fall' if x>8 else'Summer' if x>5 else 'Spring' if x>2 else 'Winter')
```

```python
df3.groupby('race_season')['average_speed'].agg(['mean' , 'count']).sort_values('mean' , ascending = False)
```
